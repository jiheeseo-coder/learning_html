```python
index = range(df2.shape[0])
df2.index = index
display(df2)
```

```python
temp_list = []
rain_list = []
day_list = []
count = 0
# for row in df2.itertuples():
#     for row2 in dfw.itertuples():
#         #print('----------------------------')
#         if row[1] != row2[3]:
#             continue
#             #break
#         #print(count," : ",row[1],", ",row2[4])
#         temp_list.append(row2[4])
#         rain_list.append(row2[5])
#         day_list.append(row2[3])
#         count +=1
        #print('어디서 누락이 됬느냐', row2[3])

for i1 in df2.index:
    for i2 in dfw.index:
        if df2['날짜'][i1] != dfw['일시'][i2]:
            continue
        temp_list.append(dfw['평균기온(°C)'][i2])
        rain_list.append(dfw['일강수량(mm)'][i2])
        day_list.append(dfw['일시'][i2])
        count +=1


print('temp_list', temp_list)
print(len(temp_list))
print('rain_list', rain_list)
print(len(rain_list))
```

```python
dd=np.array(df2['날짜'].tolist())
dl=np.array(day_list)
intersection = np.array(list(set(dd)-set(dl)))
print(intersection)
print(len(intersection))

#13*4 해서 누락된 52개의 데이터를 확인함.
```

#### 모델링

```python
from sklearn.model_selection import train_test_split
check_columns = ['평균기온', '강수량']
X = adf[check_columns]
y = adf['평균가격']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn import linear_model

lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

pred_test = lr.predict(X_test)
```

```python
#계수, 기울기 첫번째는 평균기온, 두번재는 강수량
print(lr.coef_)
```

```python
# 성능 평가
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
#ㅋㅋㅋㅋ 그래,, 앞에서 데이터 확인해볼때 부터 알아봤다,,,
# 1이랑 하나도 안가까움ㅋㅋㅋ
```
