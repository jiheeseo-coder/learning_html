##### (3) "weather202009.csv"를 data 폴더에 저장하여 weather 데이터프레임으로 로딩

```python
weather = pd.read_csv('data/weather202009.csv', encoding='CP949')
display(weather)
```

```python
print("shape\n", weather.shape)
print("\nindex\n", weather.index)
print("\ncolums\n", weather.columns)
print("\ntype\n", weather.dtypes)
```

##### (5) 기상 데이터(weather)의 상위, 하위 5개씩 확인

##### (6) 기상 데이터(weather)의 요약 통계 값 확인

```python
display(weather.describe())
```

##### (7) Weather 데이터로부터  일자 별 지점명의 평균기온으로 테이블을 재구성하여 weather2로 저장 후 출력

```python
weather2 = weather.pivot(index = '일시', columns='지점명', values='평균기온')
display(weather2)
```

##### (8) 기상 데이터(weather)의 '지점명', '평균기온', '최저기온', '최고기온', '평균 상대습도' 열을 이용하고, '일시' 열을 index로 설정하여 weather3 생성 후 확인

```python
weather3 = weather[['지점명', '평균기온', '최저기온', '최고기온', '평균 상대습도']].copy()
weather3.index = weather['일시']
display(weather3)
```

##### (9) weather3의 ‘지점명’을 범주형으로 변경 후 확인

```python
weather3['지점명'] = weather3['지점명'].astype("category")
#display(weather3)

print(weather3.dtypes)
```

##### (10) weather3의 지점별 9월 전체 평균기온, 평균습도 확인 (지점별 그룹 처리)

```python
grouped_loca = weather3.groupby('지점명')
for name, group in grouped_loca:
    print("name : ",name)
    print("groupdata : ")
    display(group)
    print()
    
display(grouped_loca[['평균기온', '평균 상대습도']].mean())
```

##### (11) weather3에 ‘일교차‘ 열을 추가하고 ‘감기지수’열을 아래 조건으로 추가하고 확인

 <일교차를 이용한 감기지수 조건>  

   https://www.weather.go.kr/weather/lifenindustry/li_asset/HELP/basic/help_02_08.jsp

  일교차               감기지수     

-------- ---------------------

  17.8이상            매우높음   

  17.8미만 14이상   높음         

  14미만 10.1이상   보통         

  10.1 미만            낮음        

```python
weather3['일교차'] = weather3['최고기온'] - weather3['최저기온']

conditionlist = [(weather3['일교차'] >= 17.8),
                 (weather3['일교차'] < 17.8) & (weather3['일교차'] >= 14),
                 (weather3['일교차'] < 14) & (weather3['일교차'] >= 10.1),
                 (weather3['일교차'] < 10.1)]
choicelist = ['매우높음', '높음', '보통', '낮음']
weather3['감기지수'] = np.select(conditionlist, choicelist, default='Not')
    
display(weather3)
```

##### (12) weather3의 ‘감기지수’ 열을 범주형으로 변경 후 확인

```python
weather3['감기지수'] = weather3['감기지수'].astype("category")

print(weather3.dtypes)
```

##### (13) weather3를 ‘감기지수’로 정렬하여 출력

```python
display(weather3.sort_values(by='감기지수'))
```

##### (14) weather3를 ‘감기지수’로 그룹화하여 ‘평균기온’,’평균 상대습도’에 대한 중앙값, 평균값, 표준편차 확인

```python
grouped_cold = weather3.groupby('감기지수')
for name, group in grouped_cold:
    print("name : ", name)
    print("groupdata : ")
    display(group)
    print()
    
display(grouped_cold[['평균기온', '평균 상대습도']].agg(['sum', 'mean', 'std']))
```
