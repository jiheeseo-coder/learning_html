#### 전력 사용패턴 예
- 지역별 전력 판매량 데이터를 가지고 클러스터링 하는 예를 소개함. 전력 사용 데이터를 읽고 서비스업과 제조업 분야의 전력 판매량을 두 개의 특성으로 2차원 평면에 지역을 나타내는 코드는 아래와 같다.

```python
# 데이터
import platform
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
%matplotlib inline
```

```python
# 한글 출력
# 구글링함: https://velog.io/@seonj102/jupyter-notebook-%EA%B7%B8%EB%9E%98%ED%94%84-%ED%95%9C%EA%B8%80-%ED%8F%B0%ED%8A%B8-%EC%84%A4%EC%A0%95
platform.system()

if platform.system() == 'Darwin': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')

plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정


# 글씨 선명하게 출력하는 설정
%config InlineBackend.figure_format = 'retina'
```

```python
#power_data = pd.read_excel('./JUNGOL/data/시도별_용도별.xls')
power_data = pd.read_excel('C:/Users/서지희/OneDrive - 강원대학교/대학/WorkSpace/Python/JUNGOL/data/시도별_용도별.xls')
print(power_data.shape)
```

```python
#구분을 인덱스로 지정
power_data = power_data.set_index('구분')
# 합계, 개성, 경기, 서울 삭제
power_data = power_data.drop(['합계', '개성', '경기', '서울'], errors='ignore')
# 합계 열도 삭제
power_data = power_data.drop('합계', axis=1)
power = power_data[['서비스업','제조업']]
#power는 서비스업과 제조업만 보여줘줘power.head()
```

##### 덴드로그램

```python
from scipy.cluster.hierarchy import dendrogram, linkage
#덴드로그램을 사용할거에요

# 그래프 사이즈 정해주기
plt.figure(figsize=(10, 5))
#덴드로그램에 사용되는 link_dist 설정
link_dist = linkage(power, metric='euclidean', method='ward')
# 덴드로그램 함수 사용할건데, link_dist 사용하고 레이블은 power.index로 지정(구분)
dendrogram(link_dist, labels = power.index)
#그래프를 보여줘라
plt.show()
```

##### 클러스터링

```python
from sklearn.cluster import KMeans
# 사이킷런의 클러스터에서 KMeans 임포트해줘

kmeans = KMeans(n_clusters=4).fit(power)
# 위에서 임포트한 KMeans 함수 쓸건데, Knn의 클러스터는 4로 해주고 power라는 데이터를 사용할거야

print(kmeans.n_clusters)
# 클러스터의 갯수 출력하기 아까 우리가 정했던것 처럼 4가 나오겠죠
```

```python
centers = kmeans.cluster_centers_
# 센터는 이걸로 해줘
print(kmeans.labels_)
#레이블 한번 출력해보기(정답)
```

```python
power.loc[:, '클러스터'] = kmeans.labels_
# 뒤에 클러스터라는 항목을 추가하는데 거기에는 레이블이 들어감
power.head()
# 5개 출력해보자
```

```python
# 그래프 그리기 위한 준비

my_markers = ['*', '^', 'o', 'v']
#모양은 이런걸 쓸거야
my_color = ['y', 'b', 'g', 'k']
# 색깔은 이걸 쓸거야, 노란색, 검은색, 초록색,, k,,?

plt.figure(figsize=(10, 8))
# 그래프 사이즈 정해주기
plt.xlabel('서비스업')
# x축 이름 지정
plt.ylabel('제조업')
# y축 이름 지정

# 그래프 그리는 반복문
for n in range(power.shape[0]):
    #print(n), 여기서 n은 강원부터 충북까지 15개
    label = kmeans.labels_[n]
    # 레이블 지정
    plt.scatter(power['서비스업'][n], power['제조업'][n], c = my_color[label], marker=my_markers[label], s = 100)
    # 그래프 그리기(X데이터: 서비스업의 도시, Y데이터: 제조업과 도시, c: 색깔(레이블별로), marker: 모양, s: 마커 크기
    plt.text(power['서비스업'][n]*1.03, power['제조업'][n]*0.98, power.index[n])
    # 인덱스 붙이기

for i in range(kmeans.n_clusters):
    plt.scatter(centers[i][0], centers[i][1], c = 'r', s = 50)
    #빨간 점 찍기
```

##### 스케일링

```python
from sklearn.preprocessing import StandardScaler
# 사이킷런의 preprocessing에서 StandardScaler라는 함수를 쓸꺼야
import warnings
warnings.simplefilter('ignore')
# 예외에 관련된 것
#power = power.fillna(power.median())
scaler = StandardScaler()

power[:] = scaler.fit_transform(power[:])
# [:] 이게 슬라이스 전체를 가져오는건데 왜 이렇게 가져오는지 모르겠슴,,, 그냥 power라고 적어도 될텐데,,,
```

##### 스케일링 이후의 덴드로그램

```python
Z = linkage(power, metric='euclidean', method='ward')
#유클리드 거리를 이용해 linkage matrix를 생성

plt.figure(figsize=(10, 5))
# 사이즈 설정
plt.title('덴드로그램')
# 덴드로그램 타이틀 설정
dendrogram(Z, labels = power.index)
# 덴드로그램 이렇게 만듭니다!
```

##### 클러스터링

```python
k = KMeans(n_clusters=4).fit(power)
# KMeans 모델을 만듬
centers = k.cluster_centers_
# 클러스터의 센터 정하기

plt.clf()
# 생성한 그래프를 clear 함
plt.figure(figsize=(10, 8))
plt.xlabel('서비스업')
plt.ylabel('제조업')

for n in range(power.shape[0]):
    label = k.labels_[n]
    plt.scatter(power['서비스업'][n], power['제조업'][n], c = my_color[label], marker=my_markers[label], s = 100)
    plt.text(power['서비스업'][n]*1.05, power['제조업'][n]*0.99, power.index[n])

for i in range(k.n_clusters):
    plt.scatter(centers[i][0], centers[i][1], c= 'r', s = 50)
```

##### 모든 특성을 고려

```python
power3 = power_data.drop(['업무용합계', '산업용합계'], axis = 1, errors = 'ignore')
power[:] = scaler.fit_transform(power3[:])

Z= linkage(power3, metric = 'euclidean', method='ward')
#여기부터는아깔 ㅇ똑같은데 모든 특성을 고려한 것임
```
